```python

# Suppose we need to return last 4 digits of the credit card number, but we dont have any built in dunction for it
# so we have to create a custom function and cast it to the .apply() method
#defining custom function which returns the last 4 digits of number
def last_four(num):
    return str(num)[-4:]
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
def quality(total_bill, tip):
    if total_bill/tip > 0.25:
        return 'Generous !!'
    else:
        return 'Other'
```

```python
quality(16.99 , 1.01)
```

```python
df['Quality'] = df[['total_bill','tip']].apply(lambda df : quality(df['total_bill'], df['tip']),axis = 1)
```

```python
df["Quality"]
```

```python
df["Quality"] = np.vectorize(quality )(df['total_bill' ], df['tip'])
```
